```python
import requests
import time
import matplotlib.pyplot as plt
```

```python
url = "https://api.wheretheiss.at/v1/satellites/25544"

latitudes = []
longitudes = []
timestamps = []

for i in range(100):
    response = requests.get(url)
    data = response.json()
    
    lat = data["latitude"]
    lon = data["longitude"]
    ts = data["timestamp"]
    
    latitudes.append(lat)
    longitudes.append(lon)
    timestamps.append(ts)
    
    print(f"{i+1}/100 - Lat: {lat:.2f}, Lon: {lon:.2f}, Time: {ts}")
    
    time.sleep(10)  

plt.figure(figsize=(12,6))
plt.plot(longitudes, latitudes, marker="o", linestyle="-", markersize=3, color="blue")
plt.title("ISS Path Over ~17 Minutes (100 points)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
  
```
